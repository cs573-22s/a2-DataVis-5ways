```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cars-sample.csv")
```

```python
df
```

```python
plt.figure(figsize=(8, 6), dpi=80)
sns.set_style("darkgrid")
sns.scatterplot(x ='Weight', y= 'MPG',hue ='Manufacturer',size = 'Weight',
                palette = ['tab:olive','tab:pink','tab:red','tab:green','tab:cyan'],alpha = 0.4, data = df,
                size_norm =(1500,5000),sizes= (30,350),legend = 'brief')

plt.legend(bbox_to_anchor = (1.02,0.8),loc ='upper left',borderaxespad =0,fontsize = 'large',frameon = False,
           labelspacing = 0.55)
plt.xticks([2000,3000,4000,5000])
plt.yticks([10,20,30,40])


plt.show()
```
