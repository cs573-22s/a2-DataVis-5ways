# Seaborn Visualization


```python
# Load in packages for all plots
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load in car data

cars = pd.read_csv("cars-sample.csv")
```

```python
# Recreate the plot
# make them all the right style
sns.set(style="darkgrid")

# generate the plot and legend off to the side
v = sns.scatterplot(data = cars, x = "Weight", y = "MPG", hue = "Manufacturer", size = "Weight", alpha = .5)
v.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1) 

# Set axis tick marks and display proper grid lines
v.set_yticks([10, 20, 30, 40])
v.set_yticks([15, 25, 35, 45], minor = True)

v.set_xticks([2000,4000,3000,5000])
v.set_xticks([1500,2500,3500,4500, 5200], minor = True)

v.grid(b = True, which = "major", linewidth = 1.25)
v.grid(b = True, which = "minor", linewidth = .75)
v.tick_params(color = 'black', bottom = True)
v.tick_params(color = 'black', left = True)
plt.xlabel("Weight", weight = "bold")
plt.ylabel("MPG", weight = "bold")

# actually show the plot
plt.show()
```

```python
# Make it accessible-- essentially, we altered the color palette and distinguished each manufactuer by a unquie marker to make this accessible for colorblind individuals

a = sns.scatterplot(data = cars, x = "Weight", y = "MPG", hue = "Manufacturer", size = "Weight", alpha = .5, style = "Manufacturer", palette = "colorblind")
a.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1) 

# Set axis tick marks and display proper grid lines
a.set_yticks([10, 20, 30, 40])
a.set_yticks([15, 25, 35, 45], minor = True)

a.set_xticks([2000,4000,3000,5000])
a.set_xticks([1500,2500,3500,4500, 5200], minor = True)

a.grid(b = True, which = "major", linewidth = 1.25)
a.grid(b = True, which = "minor", linewidth = .75)
v.tick_params(color = 'black', bottom = True)
v.tick_params(color = 'black', left = True)
plt.xlabel("Weight", weight = "bold")
plt.ylabel("MPG", weight = "bold")

# actually show the plot
plt.show()
```
